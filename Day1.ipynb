### **Types of programming language:**
 In computer Science programming languages are categorized into **6 types** as follows:


1.    **Procedural:** It is used to specify a sequence of tasks to be performed by a computer.The procedures are also known as function which are used to structure the program into a reusable block of code. 
 **Eg:** C language, Pascal.

2.   **Object-Oriented:** It is based on the concept of objects which is an data structure that contains both data and method (encapsulated) and perform operations on the data. **Objects are instances of classes** which are the blue prints that define the structure of the objects. Object has two components namely:      

 *   **Attributes:** these are the data that the object holds i.e., name, age, size etc.
 *   **Methods** are the operations that can be performed on the data for getting the data, setting the data or performing calculations on the data.
  **Eg:** Consider a class "Person". The class "Person" could define attributes such as name, age, and address, and methods such as speak(), walk(), and run(). An object of the class "Person" would be a specific person, with their own unique name, age, and address, and would have the ability to speak, walk, and run.
  
  Example of object-oriented programming languages are:Java, Python , Ruby.

3.   **Functional:** It is based on mathematical funtions Eg: Haskell and Lisp.

4.   **Logic:** It is based on formal logic and symbloic computing (deals with manipulation and solution of mathematical expression and equation in symbloic form, rather than in numerical form).

5.  **Scripting:** It is used for automate taks, such as web page interactions, data processing, system administration. Scripting languages are usually interpreted rather than compiled.
Eg: JavaScript, Python, Ruby.

6.  **Assembly:** It is an low-level programming language, used to program on CPU directly.

### **Difference between compiler and interpreter programming lang:**
Screenshot (110).png

* A compiler translates the entire program into machine code before it is executed, whereas an interpreter reads and executes each line of code one at a time.
  
* Compiled code is generally faster to run because it has already been translated into machine code, but it cannot be modified without being recompiled. Interpreted code is slower to run because it must be translated, but it is easier to modify and debug because each line of code can be executed immediately.

* **Real time example:**
There are 2 persons where person1 is from USA and person2 is from spain

 In scenario 1(compiler):

 Both the person’s can understand both of their languages i.e English and Spanish so there is no need of a translator henceforth their communication speed is faster and in the other case (interpreter).,

 The person 1 cannot understand Spanish and the person 2 cannot understand English henceforth we need a translator who knows both English and Spanish. 
When the person 1 says it in English the translator convert it into Spanish and vise-versa which leads to time-taking when compared to the first scenario because single line is been translated at that particular period of time.
